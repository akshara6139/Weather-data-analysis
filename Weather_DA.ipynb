```python
# We can use graphs for representing instead of grouping
# Bar plot requires at one numeric data and one categorical data
# Seaborn directly does aggregation but matplotlib doesn't
import seaborn as sns
```

```python
sns.barplot(y='Month',x='Temperature (C)',data=snow,estimator=len)
```

```python
# Another method of command
# estimator = len is similar to count
sns.barplot(y=snow['Month'],x=snow['Temperature (C)'],estimator=len)
```

```python
# Converting year -- to string
sns.barplot(y=snow['Year'].astype(str),x=snow['Temperature (C)'],estimator=len)
```

```python
import ipywidgets as widgets
```

```python
c=widgets.Dropdown(
    options=['rain','snow'],
    value='rain',
    description='graph:',
    disabled=False,
)

def graph(a):
  sns.barplot(y='Month',x='temperature',data=df1[df1['Precip Type']==a],estimator=len)
  plt.xlabel(a)
  plt.show()
```

```python
# Call the func(a is taken from c)
widgets.interact(graph,a=c)
```
